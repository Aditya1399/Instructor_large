```python
import torch
import transformers
from torch import cuda,bfloat16
from transformers import LlamaTokenizer, LlamaForCausalLM, GenerationConfig, pipeline,AutoTokenizer,AutoModelForCausalLM,AutoModelForSeq2SeqLM
device=f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'

bnb_config=transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=bfloat16
)
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",use_auth_token=True)

model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf"
                                              ,device_map='auto',
                                             torch_dtype=torch.float16,
                                            )
```

```python
output_list=[]
for i in range(len(docs1)):
  input_context = "You are being provided with some input.Your task is to provide insights and summarise the input" + docs1[i]
  input_ids = tokenizer.encode(input_context, return_tensors="pt")
  output = model.generate(input_ids, max_length=700, temperature=0.3)
  output_text = tokenizer.decode(output[0], skip_special_tokens=True)
  output_list.append(output_text)


print(output_list)
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
import torch

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,

    max_new_tokens=512,
    temperature=0.3,
    top_p=0.95,
    repetition_penalty=1.15
)

local_llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain import PromptTemplate, LLMChain

# Prompt
prompt = PromptTemplate.from_template("""
You are being provided with {context}.
Your task is to summarise the details in the {context} and provide
insights in the {context}
""")

# Chain
llm_chain = LLMChain(llm=local_llm,prompt=prompt)

# RunG

```
